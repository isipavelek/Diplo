Las características utilizadas para describir a cada cliente pueden ser edad, sexo, intereses, hábitos de consumo, entre otros.

## Kmeans
Método del codo

```python
# Preparamos una lista donde vamos a ir agregando las distancias medias
Sum_of_squared_distances = []
# Fiteammos un modelo para cada numero de cluster que queremos testear
K = np.arange(2,10)
for k in K:
    # Definimos y entrenamos el modelo
    km = KMeans(n_clusters=k)
    km = km.fit(X1)
    # Calculamos la distancia media y agregamos a la lista
    distancia_total = km.inertia_
    distancia_media = np.divide(distancia_total,X1.shape[0])
    Sum_of_squared_distances.append(distancia_media)
    
plt.figure(figsize = (10,7))
plt.plot(K, Sum_of_squared_distances, lw=3)
plt.scatter(K, Sum_of_squared_distances,s=55,c='r')
plt.xlabel('#Cluster')
plt.ylabel('inertia')
plt.title('Elbow method')
plt.show()
```

Efectivamente el método del codo no muestra que el a partir de 5 clusters, la inercia decae. <br>
Veamos que dice el método de silueta.

```python
silhouette_coefficients = []

for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X1)
    score = silhouette_score(X1, kmeans.labels_)
    silhouette_coefficients.append(score)

plt.figure(figsize = (10,7))    
plt.plot(K, silhouette_coefficients, lw=3)    
plt.scatter(K, silhouette_coefficients, s=55,c='r')
plt.xticks(K)
plt.xlabel("#Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

Siluetas

```python
# Aca ponen el Dataset con el que quieren trabajar
X_std = X1
# Aca definen la lista de ks para los cuales quieren hacer un gráfico
k_list = [4, 5, 6]

for i, k in enumerate(k_list):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,6))
    
    # Run the Kmeans algorithm
    km = KMeans(n_clusters=k)
    labels = km.fit_predict(X_std)
    centroids = km.cluster_centers_

    # Get silhouette samples
    silhouette_vals = silhouette_samples(X_std, labels)

    # Silhouette plot
    y_ticks = []
    y_lower, y_upper = 0, 0
    for i, cluster in enumerate(np.unique(labels)):
        cluster_silhouette_vals = silhouette_vals[labels == cluster]
        cluster_silhouette_vals.sort()
        y_upper += len(cluster_silhouette_vals)
        ax1.barh(range(y_lower, y_upper), cluster_silhouette_vals, edgecolor='none', height=1)
        ax1.text(-0.03, (y_lower + y_upper) / 2, str(i + 1))
        y_lower += len(cluster_silhouette_vals)

    # Get the average silhouette score and plot it
    avg_score = np.mean(silhouette_vals)
    ax1.axvline(avg_score, linestyle='--', linewidth=2, color='green')
    ax1.set_yticks([])
    ax1.set_xlim([-0.1, 1])
    ax1.set_xlabel('Silhouette coefficient values')
    ax1.set_ylabel('Cluster labels')
    ax1.set_title('Silhouette plot for the various clusters', y=1.02);
    
    # Scatter plot of data colored with labels
    ax2.scatter(X_std.iloc[:, 0], X_std.iloc[:, 1], c=labels)
    ax2.scatter(centroids[:, 0], centroids[:, 1], marker='*', c='r', s=250)
    ax2.set_xlabel('Spending Score (1-100)')
    ax2.set_ylabel('Anual Income')
    ax2.set_title('Visualization of clustered data', y=1.02)
    ax2.set_aspect('equal')
    plt.tight_layout()
    plt.suptitle(f'Silhouette analysis using k = {k}',
                 fontsize=16, fontweight='semibold', y=1.05);
```

Acorde al método del codo y el score de silueta, 5 cluster parecería ser la mejor representación. <br>

Se hace un barrido sobre una grilla de _eps_ y _min_samples_ para ver cuál es la mejor configuración de hiperparámetros para DBSCAN.

```python
# En este paso tenemos que definir el epsilon y en min_samples, los parametros del método.
from sklearn.cluster import DBSCAN
from utils import plot_DBSCAN

X = np.zeros((X1.shape[0], X1.shape[1]))
X[:, 0] = X1['Spending Score (1-100)']
X[:, 1] = X1['Annual Income (k$)']

eps_grid = np.arange(6, 10, 1)
min_samples_grid = np.arange(2, 6, 1)


for min_samples_val in min_samples_grid:
    fig, axes = plt.subplots(nrows=1, ncols=len(eps_grid), figsize=(15,4))
    col_ix = row_ix = 0
    print()
    for eps_val in eps_grid:
        db = DBSCAN(eps=eps_val, min_samples=min_samples_val)
        db.fit(X1)
        
        # Consultamos las etiquetas
        labels = db.labels_

        # Number of clusters in labels, ignoring noise if present.
        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        n_noise = list(labels).count(-1)

        #print('(#clusters, #noise points): {:d} - {:d}'.format(n_clusters,n_noise))
        
        plot_DBSCAN(X,labels,db,n_clusters, axes[col_ix])
        col_ix += 1            

    fig.tight_layout()
    plt.show()
```

Como se puede observar, no hay una configuración de DBSCAN que permita obtener resultados similares a KMeans. Esto se debe fundamentalmente a la geometría del los datos utilizados para la clusterización. <br>
Bajo la premisa de que el desempeño de DBSCAN sea similar al obtenido con KMeans, la configuración adoptada para DBSCAN es: **eps=9** y **min_samples=3*.

```python
db = DBSCAN(eps=9, min_samples=3)
db.fit(X1)
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,8))
plot_DBSCAN(X,db.labels_,db,n_clusters, ax)
```

## Clustering jerárquico
Se comienza el clustering jerarquico evaluando la totalidad de los niveles.

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(distance_threshold=0, n_clusters=None)
model = model.fit(X1)
```

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
Z = hierarchy.linkage(model.children_, 'ward')
```

```python
plt.figure(figsize=(20,10))
dn = hierarchy.dendrogram(Z)
```

La base del dendograma corresponde a los puntos individuales. <br>
Las líneas horizontales simplemente muestran la unión de los clusters. <br>
La longitud de las líneas verticales representa el recorrido que hizo el algoritmo para lograr unirlo con algo. 
Por lo tanto, si la longitud de las líneas verticales es "larga", significa que el punto está lejos de los demas como para formar un cluster; mientras que si la línea vertical es "corta", el punto está cerca de otros. <br>
A trevés del dendograma (de las longitudes verticales) podríamos determinar qué punto podría ser un potencial outlier.
Si cortamos con una línea horizontal a diferentes niveles vamos a poder  seleccionar diferentes estadíos de la formación del clustering jerárquico y quedarnos con una configuración que consideremos adecuada (apuntamos a obtener el mismo resultado que KMeans.

```python
n_cluster = [12, 6, 5]
fig, axes = plt.subplots(nrows=1, ncols=len(n_cluster), figsize=(20,6))
ix = 0
for k in n_cluster:
    aggloclust = AgglomerativeClustering(n_clusters=k).fit(X1)
    
    sns.scatterplot(x=X1['Spending Score (1-100)'], 
                    y=X1['Annual Income (k$)'], 
                    s=100, 
                    hue = aggloclust.labels_,
                    palette="deep",
                    ax=axes[ix],
                    legend=False)

    axes[ix].set_title("#cluster: " + str(aggloclust.n_clusters_))
    ix += 1
    
plt.show()
```

Se toman 5 clusters para tener un punto de comparación con KMeans

```python
aggloclust = AgglomerativeClustering(n_clusters=5)
aggloclust = aggloclust.fit(X1)
aggloclust.n_clusters_
```

Se agregan las etiquetas del clustering aglomerativo al dataframe.

```python
df['agglomerative_labels'] = np.zeros((df.shape[0])).astype(int)
for k in range(aggloclust.n_clusters_):
    ix = np.where(aggloclust.labels_ == k)[0]
    df.loc[ix, 'agglomerative_labels'] = k
    
df.head()
```

Representación de los clusters obtenidos con clustering jerárquico y 5 clusters.

```python
plt.figure(figsize=(8,6))

sns.scatterplot(x=X1['Spending Score (1-100)'], 
                y=X1['Annual Income (k$)'], 
                s=100, 
                hue = aggloclust.labels_,
                palette="deep")
plt.title('Data points and cluster centroids')
plt.show()
```

# Reducción de dimensionalidad
Evaluamos si vale la pena aplicar reducción de dimensionalidad. <br>
En lo desarrollado hasta aquí hemos utilizado dos variables. Con esto presente, para que la aplicación de PCA valga la pena deberíamos encontrar un espacio unidimensional que explique casi el total de la varianza. A priori, parece poco probable. De todas maneras, evualamos PCA. <br>

```python
# Contruimos un DataFrame que únicamente tenga los datos del DataSet orginal.
df_pca = df.iloc[:,:-3].copy()
display_html(df_pca)
```

Instanciamos PCA sin hiperparámetros para ver las componentes principales y poder determinar qué variables son las que concetran la mayor información.

```python
from sklearn.decomposition import PCA
pca = PCA()
components = pca.fit_transform(df_pca)
```

```python
import plotly.express as px

labels = {
    str(i): f"PC {i+1} ({var:.1f}%)"
    for i, var in enumerate(pca.explained_variance_ratio_ * 100)
}

fig = px.scatter_matrix(
    components,
    labels=labels,
    dimensions=range(df_pca.shape[1])
)
fig.update_traces(diagonal_visible=False)
fig.show()
```

Notar que 2 componentes principales explican el 89.2% de la varianza. <br>

Se utiliza PCA con 2 componentes (respecto de lo que veníamos trabajando, tendríamos una transformación de variables)

```python
scaler = StandardScaler()
X1 = scaler.fit_transform(X1)
```

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X1)
```

Se instancia KMeans con **n_cluster=5** con el objetivo de poder comparar los resultados entre kmeans con y sin reducción de dimensionalidad.

```python
kmeans_pca = KMeans(n_clusters=5, random_state=0)
kmeans_pca.fit(principal_components)
```

Se representan los clusters obtenidos con PCA

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=principal_components[:, 0], 
                y=principal_components[:, 1], 
                hue = kmeans_pca.labels_,
                palette='rainbow')
sns.scatterplot(x=kmeans_pca.cluster_centers_[:, 0], 
                y=kmeans_pca.cluster_centers_[:, 1],
                color='black', 
                marker="^", s=500)
plt.title('Data points and cluster centroids')
plt.show()
```

```python
labels = {
    str(i): f"PC {i+1} ({var:.1f}%)"
    for i, var in enumerate(pca.explained_variance_ratio_ * 100)
}

fig = px.scatter_matrix(
    principal_components,
    labels=labels,
    dimensions=range(2)
)
fig.update_traces(diagonal_visible=False)
fig.show()
```

Se agregan las etiquetas de kmeans_pca al _DataFrame_ final

```python
df['kmeans_pca_labels'] = np.zeros((df.shape[0])).astype(int)
for k in range(kmeans_pca.n_clusters):
    ix = np.where(kmeans_pca.labels_ == k)[0]
    df.loc[ix, 'kmeans_pca_labels'] = k
    
df.head()
```
